```python
data_sel_char1 = data_group[['count']]
data_sel_char1

data_normalized1 = data_sel_char1.copy()
data_normalized1['count_n'] = (data_sel_char1['count'] - data_sel_char1['count'].min()) / (data_sel_char1['count'].max() - data_sel_char1['count'].min())
data_normalized1
```

```python

import pandas as pd
import json
import re
from sklearn.pipeline import Pipeline
from sklearn.ensemble import IsolationForest
import numpy as np
import dask.dataframe as dd


pcap_data = pd.read_csv('/home/alexfrancow/AAA/ransomware2s.csv')
pcap_data.columns = ['no', 'time', 'AA', 'A', 'ipsrc', 'ipdst', 'proto', 'len']
pcap_data['info'] = "null"
pcap_data.parse_dates=["time"]

pcap_data['num'] = 1
df = pcap_data
df

```

#### Convertimos la columna time en datetime y en indice
#### Agrupamos por ipdst, proto y data y hacemos un count que irá en una columna nueva llamada count
#### Falta cambiar la columna indice por frequencias de 5 segundos

```python

df['time'] = pd.to_datetime(df['time'])
df.index = df['time']
data = df.copy()


data_group = pd.DataFrame({'count': data.groupby(['ipdst', 'proto', data.index]).size()}).reset_index()
#data_group['count'] = data_group['count'].astype(float)
pd.options.display.float_format = '{:,.0f}'.format
data_group.index = data_group['time']
data_group

data_group2 = data_group.groupby(['ipdst','proto']).resample('5S', on='time').sum().reset_index().dropna()
data_group2
#datafreq1 = data_group.resample("5s")['count'].agg(['sum']).reset_index()
#datafreq1 = data_group.resample("5s").sum()
#datafreq1 = data_group.resample("5s").mean().reset_index()
#datafreq1

data_group3 = data_group2.groupby(['time']).sum().reset_index().dropna()
data_group3
```

```python
data_sel_char1 = data_group2[['count']]
data_sel_char1

data_normalized1 = data_sel_char1.copy()
data_normalized1['count_n'] = (data_sel_char1['count'] - data_sel_char1['count'].min()) / (data_sel_char1['count'].max() - data_sel_char1['count'].min())
data_normalized1
```

```python
datatrain = data_normalized1.iloc[100:110000]
datatrain
iforest = IsolationForest(n_estimators=100, contamination=0.00001, max_samples=256)
iforest.fit(datatrain)

prediction = iforest.predict(data_normalized1)

data_group2['prediction'] = prediction
data_group3['prediction'] = prediction
data_group2
```
